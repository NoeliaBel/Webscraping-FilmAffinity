```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import re
```

# Webscraping of the Spanish Movie Listings on Filmaffinity

```python
url = "https://www.filmaffinity.com/es/topcat.php?id=new_th_es"
```

```python
response = requests.get(url)
print(bool(response))
soup = BeautifulSoup(response.text, "html.parser")
```

### Creation of the dataframe with the characteristics of the movies
- Duration
- Position
- Category
- Director
- Cast
- Rating
- Number of votes

```python
soup
```

```python
soup.find("h3")
```

```python
movies = soup.find_all("div", class_="top-movie")
# I can use instead of class_, attrs = {'class':"top-movie"}  )
```

```python
movies[0].find("h3").text
```

```python
title=[]
for i in range(40): # Title of each movie
    title.append(movies[i].find("h3").text)
    print(movies[i].find("h3").text)
```

```python
duration=[]
for i in range(40): # Duration of each movie
    duration.append(movies[i].find("div", class_="duration").text.split()[0])
    print(movies[i].find("div", class_="duration").text.split()[0])
```

```python
position=[]
for i in range(40): # Position of each movie
    position.append(movies[i].find("div", class_="pos").text)
    print(movies[i].find("div", class_="pos").text)
```

```python
category=[]
for i in range(40): # Category of each movie
    category.append(movies[i].find("div", class_="synop").find("a").text)
    print(movies[i].find("div", class_="synop").find("a").text)
```

```python
director=[]
for i in range(40): # Director of each movie
    director.append(movies[i].find("div", class_="director").text)
    print(movies[i].find("div", class_="director").text)
```

```python
cast=[]
for i in range(40): # Cast of each movie
    cast.append(movies[i].find("div", class_="cast").text)
    print(movies[i].find("div", class_="cast").text)
```

```python
rating=[]
for i in range(40): # Rating of each movie
    rating.append(movies[i].find("div", class_="avg-rating").text)
    print(movies[i].find("div", class_="avg-rating").text)
```

```python
votes=[]
for i in range(40): # Number of votes of each movie
    votes.append(movies[i].find("div", class_="rat-count").text.split()[0])
    print(movies[i].find("div", class_="rat-count").text.split()[0])
```

```python
df = pd.DataFrame()
df["Title"]=title
df["Duration"]=duration
df["Position"]=position
df["Category"]=category
df["Director"]=director
df["Cast"]=cast
df["Rating"]=rating
df["Votes"]=votes
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import re
```

# Webscraping of the Spanish Movie Listings on Filmaffinity

```python
url = "https://www.filmaffinity.com/es/topcat.php?id=new_th_es"
```

```python
response = requests.get(url)
print(bool(response))
soup = BeautifulSoup(response.text, "html.parser")
```

### Creation of the dataframe with the characteristics of the movies
- Duration
- Position
- Category
- Director
- Cast
- Rating
- Number of votes
